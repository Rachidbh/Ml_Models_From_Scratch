<h1 style="text-align:center;font-family:serif;font-size:3em;margin:30px">Polynomial regression<h1>

In this notebook, we will develop a Polynomial regression model.

**Import the necessary libraries**

```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

**Build the model**

```python
def model(X,teta):
    return X.dot(teta)

def loss_function(X,y,teta):
    m=len(y)
    return (1/2*m)*np.sum((model(X,teta)-y)**2)
def grad(X,y,teta):
    m=len(y)
    return (1/m)*X.T.dot(model(X,teta)-y)
def gradient_descent(X,y,teta,learning_rate,n_iterations):
    loss=[]
    for i in range(n_iterations):
        teta=teta-learning_rate*grad(X,y,teta)
        loss.append(loss_function(X,y,teta))
    plt.plot(loss)
    plt.xlabel("iteration")
    plt.ylabel("loss function")
    return teta

def EvaluationModel(y_true ,y_pred):
    mae = np.mean(np.abs(y_true - y_pred))
    mse = np.mean((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2) 
    ss_res = np.sum((y_true - y_pred) ** 2) 
    r2 = 1 - (ss_res / ss_tot)
    print("MAE = ",mae)
    print("MSE = ",mse)
    print("R2 = ",r2)
```

**Load the data**

```python
x,y=make_regression(n_samples=100,n_features=1,noise=10,random_state=10)
y=y+abs(y/2)
y=y.reshape(y.shape[0],1)
print("shape of X",x.shape)
print("shape of y",y.shape)
```

**Plot the dataset**

```python
plt.scatter(x,y)
plt.title("Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

We will perform a variable transformation by creating additional features from the column x. We will add a column for x² and then apply the linear regression algorithm.

![Alt Text](datasets/rp.png)

```python
X=np.hstack((x,np.ones(x.shape)))
X=np.hstack((x**2,X))
print("shape of X",X.shape)
print("shape of y",y.shape)
```

**Initialize the model parameters**

```python
teta=np.random.randn(3,1)
teta
```

```python
plt.scatter(x,y)
plt.scatter(x,model(X,teta),c='r')
plt.title("Untrained model")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Train the model**

```python
teta=gradient_descent(X,y,teta,learning_rate=0.1,n_iterations=100)
```

**Plot the trained model**

```python
plt.scatter(x,y)
plt.scatter(x,model(X,teta),c='r')
plt.title("Trained model")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Evaluate the performance of the model**

```python
y_pred=model(X,teta)
EvaluationModel( y,y_pred)
```
